```python
import pandas as pd
import numpy as np
import ast
import gzip
import json
import matplotlib.pyplot as plt
import seaborn as sns
# import nltk
# from nltk.sentiment.vader import SentimentIntensityAnalyzer
# import re
```

```python
###Asignamos los paths a variables para leer los archivos
steam_games_path = "drive/MyDrive/Henry_Proyecto_Individual_1/Datasets/steam_games.json.gz"
user_reviews_path = "drive/MyDrive/Henry_Proyecto_Individual_1/Datasets/user_reviews.json.gz"
users_items_path = "drive/MyDrive/Henry_Proyecto_Individual_1/Datasets/users_items.json.gz"
```

```python
###Convertimos el archivo json comprimido con gzip a un dataframe de pandas
steam_games = pd.read_json(steam_games_path, compression="gzip", lines=True)
steam_games
```

```python
###Convertimos el user_review un dataframe
# user_review = pd.read_json(user_reviews_path, compression="gzip", lines=True)
rows=[]
with gzip.open(user_reviews_path, 'rt', encoding='MacRoman') as archivo:
    for line in archivo.readlines():
        rows.append(ast.literal_eval(line))

user_review = pd.DataFrame(rows)
user_review
```

```python
user_review
```

```python
# user_items = pd.read_json(users_items_path, compression="gzip", lines=True)
rows = []
with gzip.open(users_items_path, 'rt', encoding='MacRoman') as archivo:
  for line in archivo.readlines():
      rows.append(ast.literal_eval(line))

user_items = pd.DataFrame(rows)
user_items = user_items.explode('items').reset_index()
user_items = user_items.drop(columns="index")
user_items = pd.concat([user_items, pd.json_normalize(user_items['items'])], axis=1)
user_items = user_items.drop(columns='items')
user_items
```

###Convertimos los data frames en archivos csv para poderlo manejar de mejor manera

```python
# Guardar el DataFrame en un archivo CSV sin incluir los índices
steam_games.to_csv("drive/MyDrive/Henry_Proyecto_Individual_1/Datasets/CSV/steam_games.csv", index=False)
user_review.to_csv("drive/MyDrive/Henry_Proyecto_Individual_1/Datasets/CSV/user_review.csv", index=False)
user_items.to_csv("drive/MyDrive/Henry_Proyecto_Individual_1/Datasets/CSV/user_items.csv", index=False)
```
